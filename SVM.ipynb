$w = [w_0, w_1, w_2]^T $

$a(x) = sign(w^T * x) $

Вычислите количество неверных классификаций в обучающей выборке по формуле:

$$Q(a,X) = \sum_ {i=1}^n [a(x_i) \not = y_i] =  \sum_ {i=1}^n [sign(w^T*x_i) \not = y_i]$$
